```python
import numpy as np
import matplotlib.pyplot as plt

def plot_vectors(a, b, projection_ab=None, title='Proyección de vectores'):
    """Función para visualizar los vectores."""
    fig, ax = plt.subplots(figsize=(8, 8))

    # Límites del gráfico
    max_val = max(np.max(np.abs(a)), np.max(np.abs(b))) * 1.2
    ax.set_xlim([-max_val, max_val])
    ax.set_ylim([-max_val, max_val])
    ax.set_aspect('equal')
    ax.grid(True)
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title(title)

    # Origen
    origin = np.array([0, 0])

    # Dibujar vectores
    ax.quiver(*origin, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector a')
    ax.quiver(*origin, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='g', label='Vector b')

    # Dibujar la proyección si se proporciona
    if projection_ab is not None:
        # Cambiado linestyle a un valor seguro
        ax.quiver(*origin, projection_ab[0], projection_ab[1], angles='xy', scale_units='xy', scale=1, color='r', linestyle='solid', label='Proyección de a sobre b')
        # Dibujar la línea de error (ortogonal)
        error_vector = a - projection_ab
        # Cambiado linestyle a un valor seguro
        ax.quiver(projection_ab[0], projection_ab[1], error_vector[0], error_vector[1], angles='xy', scale_units='xy', scale=1, color='purple', linestyle='solid', label='Vector de error (a - proyección)')

    ax.legend()
    plt.show()

# --- Definición de vectores ---
a = np.array([4, 6])
b = np.array([8, 2])

# ====================================================================
#              1. Solución Analítica
# ====================================================================

print("=== Solución Analítica ===")
# Calcular el producto punto
dot_product = np.dot(a, b)          #a*b
# Calcular la norma al cuadrado de b
norm_b_squared = np.dot(b, b)       #//b//^2
# Calcular el escalar de la proyección
scalar = dot_product / norm_b_squared
# Calcular el vector de proyección
projection_a_on_b_analytical = scalar * b

print(f"Producto punto (a · b): {dot_product}")
print(f"Norma de b al cuadrado (||b||^2): {norm_b_squared}")
print(f"Escalar (c): {scalar:.4f}")
print(f"Proyección de a sobre b (analítica): {projection_a_on_b_analytical}")

# Visualizar la solución analítica
plot_vectors(a, b, projection_a_on_b_analytical, 'Proyección analítica')


# ====================================================================
#              2. Solución Iterativa (Gradiente Descendente)
# ====================================================================

print("\n=== Solución Iterativa (Gradiente Descendente) ===")
# Parámetros para el descenso de gradiente
learning_rate = 0.01  # Tasa de aprendizaje
epochs = 500         # Número de iteraciones
c = 0.0              # Inicializamos el escalar 'c'

# Bucle de optimización
cost_history = []
for i in range(epochs):
    # Calculamos la función de costo
    cost = np.sum((a - c * b)**2)
    cost_history.append(cost)

    # Calculamos el gradiente
    gradient = -2 * np.dot(a - c * b, b)

    # Actualizamos el escalar 'c'
    c = c - learning_rate * gradient

    if i % 50 == 0:
        print(f"Época {i}: c = {c:.4f}, Costo = {cost:.4f}")

projection_a_on_b_iterative = c * b
print(f"\nValor final de c: {c:.4f}")
print(f"Proyección de a sobre b (iterativa): {projection_a_on_b_iterative}")

# Visualizar la solución iterativa
plot_vectors(a, b, projection_a_on_b_iterative, 'Proyección iterativa (Gradiente descendente)')

# Visualizar la curva de costo para entender la optimización
plt.figure()
plt.plot(cost_history)
plt.title('Curva de Costo del Gradiente Descendente')
plt.xlabel('Época')
plt.ylabel('Costo (error)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt

def load_and_preprocess_image(filepath, size=(128, 128)):
    """
    Carga una imagen, la convierte a escala de grises, la redimensiona
    y la normaliza a un rango de 0 a 1.
    """
    img = cv2.imread(filepath, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise FileNotFoundError(f"Error: No se pudo cargar la imagen en la ruta: {filepath}")
    img_resized = cv2.resize(img, size, interpolation=cv2.INTER_AREA)
    img_normalized = img_resized / 255.0
    return img_normalized

# ====================================================================
#              0. Cargar y Preprocesar Imágenes
# ====================================================================

# Asegúrate de que las imágenes estén en una carpeta llamada 'imagenes'
# en el mismo directorio que este script.
image_folder = 'imagenes'
image_size = (128, 128)

image_path_a = os.path.join(image_folder, '/content/1.jpg')
image_path_b = os.path.join(image_folder, '/content/2.jpg')

try:
    img_a = load_and_preprocess_image(image_path_a, image_size)
    img_b = load_and_preprocess_image(image_path_b, image_size)
except FileNotFoundError as e:
    print(e)
    print("Por favor, asegúrate de que el directorio 'imagenes' y los archivos 'image_a.jpg' y 'image_b.jpg' existan.")
    exit()

# Convertir imágenes a vectores 1D
vec_a = img_a.flatten()
vec_b = img_b.flatten()

print(f"Dimensiones de la Imagen A: {img_a.shape}")
print(f"Longitud del Vector A: {len(vec_a)}")
print(f"Dimensiones de la Imagen B: {img_b.shape}")
print(f"Longitud del Vector B: {len(vec_b)}")

# ====================================================================
#              1. Solución Analítica para Imágenes
# ====================================================================

print("\n=== Solución Analítica (Imágenes) ===")
# Calcular el producto punto
dot_product_img = np.dot(vec_a, vec_b)          #mide similitud entre las dos imagenes
# Calcular la norma al cuadrado de vec_b
norm_b_squared_img = np.dot(vec_b, vec_b)

if norm_b_squared_img == 0:
    scalar_img_analytical = 0
else:
    scalar_img_analytical = dot_product_img / norm_b_squared_img

print(f"Producto punto (vec_a · vec_b): {dot_product_img:.4f}")
print(f"Norma de vec_b al cuadrado (||vec_b||^2): {norm_b_squared_img:.4f}")
print(f"Escalar (c) analítico: {scalar_img_analytical:.4f}")

# ====================================================================
#              2. Solución Iterativa (Gradiente Descendente para Imágenes)
# ====================================================================

print("\n=== Solución Iterativa (Gradiente Descendente para Imágenes) ===")
# Parámetros para el descenso de gradiente
learning_rate_img = 0.000001
epochs_img = 5000
c_img = 0.0              # Inicializamos el escalar 'c'

cost_history_img = []
for i in range(epochs_img):
    # Función de costo
    cost_img = np.sum((vec_a - c_img * vec_b)**2)    #mide que tan bien c*b se aproxima a "a"
    cost_history_img.append(cost_img)

    # Gradiente
    gradient_img = -2 * np.dot(vec_a - c_img * vec_b, vec_b)

    # Actualizamos el escalar 'c'
    c_img -= learning_rate_img * gradient_img

    if i % 500 == 0:
        print(f"Época {i}: c = {c_img:.6f}, Costo = {cost_img:.4f}")

print(f"\nValor final de c (iterativo): {c_img:.4f}")

# ====================================================================
#              3. Gráfica de la Curva de Costo
# ====================================================================

plt.figure(figsize=(10, 5))
plt.plot(cost_history_img)
plt.title('Curva de Costo del Gradiente Descendente para Imágenes')
plt.xlabel('Época')
plt.ylabel('Costo (error)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# ============================
# 1. Definir vectores de ejemplo
# ============================
x_n = np.array([2.0, -1.0, 0.5])
x_m = np.array([1.0, 0.5, -0.5])

# ============================
# 2. Funciones de pérdida y gradiente
# ============================
def loss(alpha):
    """Función de pérdida: L(alpha) = ||x_n - alpha * x_m||^2"""
    return np.sum((x_n - alpha * x_m)**2)

def grad(alpha):
    """Gradiente: dL/dalpha = -2 * x_m^T (x_n - alpha x_m)"""
    return -2.0 * np.dot(x_m, (x_n - alpha * x_m))

# ============================
# 3. Solución analítica
# ============================
alpha_star = np.dot(x_m, x_n) / np.dot(x_m, x_m)

# ============================
# 4. Descenso por gradiente
# ============================
eta = 0.1       # tasa de aprendizaje
steps = 12      # número de iteraciones
alpha0 = 0.0    # valor inicial

alphas = [alpha0]
losses = [loss(alpha0)]
grads = [grad(alpha0)]

alpha = alpha0
for t in range(1, steps+1):
    g = grad(alpha)
    alpha = alpha - eta * g
    alphas.append(alpha)
    losses.append(loss(alpha))
    grads.append(g)

# Guardar trayectoria en DataFrame
df = pd.DataFrame({
    "step": list(range(0, steps+1)),
    "alpha": alphas,
    "loss": losses,
    "grad": grads
})
print("Trayectoria descenso por gradiente:")
print(df)

# ============================
# 5. Graficar parabola y pasos
# ============================
alpha_range = np.linspace(alpha_star - 2.0, alpha_star + 2.0, 400)
loss_curve = [loss(a) for a in alpha_range]

plt.figure(figsize=(8,5))
plt.plot(alpha_range, loss_curve, label="L(alpha) = ||x_n - alpha x_m||^2")
plt.scatter(alphas, losses, color="red", zorder=5)

# Dibujar flechas entre los puntos
for i in range(len(alphas)-1):
    plt.annotate("",
                 xy=(alphas[i+1], losses[i+1]),
                 xytext=(alphas[i], losses[i]),
                 arrowprops=dict(arrowstyle="->", lw=1))

plt.axvline(alpha_star, linestyle='--', label=f"alpha* = {alpha_star:.4f}")
plt.title("Descenso por gradiente sobre la parábola (pérdida L2)")
plt.xlabel("alpha")
plt.ylabel("L(alpha)")
plt.legend()
plt.grid(True)
plt.show()
```
